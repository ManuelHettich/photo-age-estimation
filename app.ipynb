```python
#|export
from fastai.vision.all import *
import gradio as gr

def get_float_labels(path):
    return float(path.name.split('_')[0])
```

```python
im = PILImage.create("10.jpg")
im.thumbnail((200,200))
im
```

```python
#|export
learn = load_learner("model_new_lr_10epochs.pkl")
```

```python
#|export
def predict_image(img):
    pred, _, _ = learn.predict(img)
    return int(pred[0])
```

```python
learn.predict(im)
```

```python
%time predict_image(im)
```

```python
#|export
image = gr.Image(shape=(200, 200))
label = gr.Label()
examples = ['10.jpg', '50.jpg', '100.jpg']

intf = gr.Interface(fn=predict_image, inputs=image, outputs=label, examples=examples)
intf.launch(inline=False)
```
